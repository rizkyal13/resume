# Pertemuan ketiga

Review sedikit tentang stacks
Tambahan ilmu ==> membuat Angka dari Desimal ke Biner

```python
#Desimal ke Biner
#Pertama Siapkan modul stacks
def stack():
    s=[]
    return (s)
def push(s,data):
    s.append(data)
def pop(s):
    data = s.pop()
    return(data)
def peek(s):
    return(s[len(s)-1])
def isEmpty(s):
    return(s==[])
def size(s):
    return(len(s))
```

```python
#Desimal ke Biner
#Eksekusi
bilangan = 5

hasilBagi=bilangan
st = stack()
while hasilBagi>0:
    temp=hasilBagi%2
    push(st,temp)
    hasilBagi=hasilBagi//2
    
temp=''
while not(isEmpty(st)):
    temp=temp+str(pop(st))
print(temp)
```

# Infix, Prefix, dan Postfix

Infix adalah model matematika yang operatornya berada didalam operan

NOTE: PADA PREFIX DAN POSTFIX, OPERAN HARUS TETAP URUT

    Contoh: 5 + 6 ('+' diantara dua buah operan)
Prefix = operatornya berada sebelum operan
    Contoh: A+B-C
    Maka, -+ABC, atau A+B = D, maka -DC
Postfix = operatornya berada setelah operan
    Contoh: A + B - C
    maka, AB + C -
    
clue: opetor tetap berada di tengah tengah operan


# Queues

Queues (antrian)

menggunakan konsep FIFO (First In First Out).
Terdiri dari dua arah. Ada Front dan juga rear
Front digunakan sebagai tempat keluar, tan rear digunakan sebagai tempat masuk




Ilustrasi Queue
